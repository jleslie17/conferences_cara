```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_excel("Downloads/conferences_cara-master/data/RegistrationBackup19Sept.xlsx")
```

```python
SH = df['SH Knowledge']
SC = df['SC Knowledge']
II = df['II Knowledge']

```

```python
def tokenizer(x):
    if isinstance(x, float):
        x = 0
    else:
        if x == 'None':
            x = 0
        if x == 'NEED INFO':
            x = 0
        else: 
            x = x.split(",")
            x = [i.strip(' ') for i in x]
    return x
  
 
```

```python
SH = SH.map(tokenizer)
SC = SC.map(tokenizer)
II = II.map(tokenizer)

```

```python
#create inverted index

from collections import defaultdict
inverted_idx = defaultdict(list)
keys_SC = set()
keys_SH = set()
keys_II = set()
def create_keys(x, keys):
    if isinstance(x, list):
        for i in x: 
            keys.add(i)

        
            
```

```python
SC.apply(create_keys, args=(keys_SC,))
SH.apply(create_keys, args=(keys_SH,))
II.apply(create_keys, args=(keys_II,))
keys_SC

```

```python
## create inverted indexes for easy cosine similarity calculations

inverted_index_SC = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_SC)
inverted_index_SH = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_SH)
inverted_index_II = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_II)
def create_index(S, keys, invertdict):
     for index, value in S.iteritems():
        if isinstance(value, list):
            for i in value:
                invertdict.set_value(index, value, 1)
     
create_index(SC, keys_SC, inverted_index_SC)
create_index(SH, keys_SH, inverted_index_SH)
create_index(II, keys_II, inverted_index_II)
```

```python
##create learnings 
SH_learn = df['SH Learnings']
SC_learn = df['SC Learnings']
II_learn = df['II Learnings']
```

```python
#tokenize learnings
SH_learn = SH_learn.map(tokenizer)
SC_learn = SC_learn.map(tokenizer)
II_learn = II_learn.map(tokenizer)
```

```python
keys_SCL = set()
keys_SHL =set()
keys_IIL=set()
SC_learn.apply(create_keys, args=(keys_SCL,))
SH_learn.apply(create_keys, args=(keys_SHL,))
II_learn.apply(create_keys, args=(keys_IIL,));
```

```python
inverted_index_IIL = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_IIL)
inverted_index_SHL = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_SHL)
inverted_index_SCL = pd.DataFrame(0, index=range(1, df.shape[0]), columns=keys_SCL)

create_index(SC_learn, keys_SCL, inverted_index_SCL)
create_index(SH_learn, keys_SHL, inverted_index_SHL)
create_index(II_learn, keys_IIL, inverted_index_IIL)
```
